```python
values = ['RELIANCE','TCS','HDFCBANK', 'INFY', 'HINDUNILVR', 'HDFC', 'ICICIBANK','KOTAKBANK','SBIN','BAJFINANCE','BHARTIARTL','ITC','HCLTECH','ASIANPAINT','WIPRO','AXISBANK','MARUTI','LT','ULTRACEMCO','DMART','ADANIGREEN','BAJAJFINSV','SUNPHARMA','ADANIPORTS','HDFCLIFE','TITAN','ONGC','HINDZINC','ADANIENT']
```

```python
from getDataFromNSE import getDataFromNSE
a = getDataFromNSE.data()

import time
times = {}
for i in values:
  print(i)
  s = time.time()
  df = a.returnsForStocks(stockSymbol=i, full_data="Yes")
  times[i] = [len(df), time.time()-s]
```

```python
times
```

```python
t = 0
for i in times:
  t += times[i][1]

print(t)

r = 0
for i in times:
    r += times[i][0]

print(t/r)
```

```python
df = pd.DataFrame.from_dict(times)
df.to_csv('Test_1.csv')
```
